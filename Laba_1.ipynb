```python
x = int(input())
y = (x*x -4*x + 4) / ((x+5)*(x+5) - 49)
print(y)
```

```python
from math import cos, sin, tan
x = int(input())
y = int(input())
w = (sin(x) + cos(y)) / (cos(x) - sin(y)) * tan(x*y)
print(w)
```

```python
from math import sqrt
katet_a = int(input())
katet_b = int(input())
square = katet_a * katet_b / 2
perimetr = katet_a + katet_b + sqrt(katet_a**2 + katet_b**2)
print(square)
print(perimetr)
```
